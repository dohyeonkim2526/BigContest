```
import pandas as pd
import numpy as np
```

```
rating= pd.read_excel('/content/drive/My Drive/2019_rating.xlsx')
rating
```

```
rating['시간대']
```

```
rating.drop(rating.tail(1).index,inplace=True)
idx_list = rating['시간대'].tolist()
print(idx_list)
```

```
del rating['2019-01-01 to 2019-12-31']
```

```
col_names = rating.columns.tolist()
col_names = col_names[1:]
```

```
new_idx = []
for col in col_names:
  for date in idx_list:
    hap = col + " " + date
    new_idx.append(hap)
```

```
new_idx = pd.DataFrame(new_idx)
```

```
len(new_idx)
```

```
df_lst = []
for i in col_names:
  df_lst.append(rating[i])

new_df = pd.concat(df_lst, ignore_index=True)
```

```
final_df = pd.concat([new_idx, new_df], axis=1)
final_df
```

```
final_df.columns = ['date','rate']
final_df = final_df.set_index('date')
final_df
```

```
idx_list = final_df.index.tolist()

## 인덱스 20개씩 묶은 리스트 만들기 

result_20 = []
for i in range(0,len(final_df),20):
    tmp = []
    for j in range(0,20):
        tmp.append(idx_list[j])
    result_20.append(tmp)
    idx_list = idx_list[20:]
```

```
tmp_list=[]
for i in range(len(result_20)):
    tmp_list.append(result_20[i][0])
```

```
cols_list = final_df.columns.tolist()
t_df = pd.DataFrame(index=range(0), columns=tmp_list)

for i in range(len(result_20)):
    tdf = pd.DataFrame(final_df, columns=cols_list, index=result_20[i])
    tmp = pd.DataFrame(tdf.mean())
    t_df.iloc[:,i] = tmp.iloc[:,0]
    
```

```
t_df
```

```
rating_df = t_df.T
```
